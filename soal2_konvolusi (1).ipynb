# Soal 2 - Konvolusi
Penerapan LPF, HPF, dan BPF pada citra grayscale.

```
# IMPORT LIBRARY
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# BUAT FOLDER OUTPUT
os.makedirs('output', exist_ok=True)

# LOAD GAMBAR
img = cv2.imread('gambar1.jpg', 0)
if img is None:
    raise ValueError("Gambar1.jpg")

# KERNEL
lpf = (1/9)*np.ones((3,3))
hpf = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
bpf = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])

# KONVOLUSI
img_lpf = cv2.filter2D(img, -1, lpf)
img_hpf = cv2.filter2D(img, -1, hpf)
img_bpf = cv2.filter2D(img, -1, bpf)

# VISUALISASI
titles = ['Original', 'LPF', 'HPF', 'BPF']
images = [img, img_lpf, img_hpf, img_bpf]

plt.figure(figsize=(10,6))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()

# NILAI RATA-RATA
print("Mean Original:", np.mean(img))
print("Mean LPF:", np.mean(img_lpf))
print("Mean HPF:", np.mean(img_hpf))
print("Mean BPF:", np.mean(img_bpf))

# SIMPAN OUTPUT
cv2.imwrite('output/original.png', img)
cv2.imwrite('output/lpf.png', img_lpf)
cv2.imwrite('output/hpf.png', img_hpf)
cv2.imwrite('output/bpf.png', img_bpf)
```
